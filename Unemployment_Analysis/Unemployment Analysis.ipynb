```python
df1 = df.drop(columns=['Region', ' Date', ' Frequency', 'Region.1'])
```

```python
plt.figure(figsize =(5,3))
sns.heatmap(df1.corr(), annot=True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.title('INDIA - UNEMPLOYMENT')
sns.histplot(x=' Estimated Employed', hue='Region.1', data=df)
plt.show()

plt.subplot(2,1,2)
sns.histplot(x=' Estimated Unemployment Rate (%)', hue='Region.1', data=df)
plt.show()
```

```python
plt.figure(figsize = (20, 5))
plt1 = sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df)
plt1.set_title('UNEMPLOYMENT RATE (STATE WISE)', size=14)
plt1.set_xlabel('REGION', fontsize=14)
plt1.set_ylabel('ESTIMATED UNEMPLOYMENT RATE(%)', fontsize=14)
plt1.set_xticklabels(plt1.get_xticklabels(), rotation=90, size=14)
plt.show()
```

```python
plt.figure(figsize = (20, 7))
plt.title('UNEMPLOYMENT RATE (STATE WISE)')
a = df.groupby(['Region'], as_index=False) [' Estimated Unemployment Rate (%)'].sum().sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
plt1 = sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=a)
plt1.set_xlabel('REGION', fontsize=14)
plt1.set_ylabel('ESTIMATED UNEMPLOYMENT RATE(%)', fontsize=14)
plt1.set_xticklabels(plt1.get_xticklabels(), rotation=90, size=14)
plt.show()
```

```python
plt.figure(figsize = (20, 7))
plt.title('UNEMPLOYMENT RATE (REGION WISE)')
a = df.groupby(['Region.1'], as_index=False) [' Estimated Unemployment Rate (%)'].sum().sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
plt1 = sns.barplot(x='Region.1', y=' Estimated Unemployment Rate (%)', data=a)
plt1.set_xlabel('REGION', fontsize=14)
plt1.set_ylabel('ESTIMATED UNEMPLOYMENT RATE(%)', fontsize=14)
plt1.set_xticklabels(plt1.get_xticklabels(), rotation=360, size=14)
plt.show()
```

# Inference:

__From the above barplots, it is evident that Estimated Unemployment Rate (%) - Statewise is higher in "HARYANA", followed by "TRIPURA", "JHARKHAND" and lowest in "MEGHALAYA".__

__Whereas in Region wise comparison, Estimated Unemployment Rate (%) is higher in "NORTH" and lowest in "WEST" region.__
